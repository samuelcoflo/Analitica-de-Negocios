0. Cargar las librerias

```python
import numpy as np #Funciones numericas transformacion y analisis de datos
import pandas as pd #Conetar sistema con la base de datos
```

1. Conexion con la base de datos

```python
nxl="/content/sample_data/0. SolicitantesCrédito(Quiz 1).xlsx"
XDB=pd.read_excel(nxl,sheet_name=0) #Dataframe: Tabla de Datos - Tablas de Word
```

2. Llamar la variable de caracterizacion (Edad)

```python
XDe=XDB[["Edad"]]
print(XDe)
XDe=XDe.dropna() #Evitar Celdas vacias
XDe=np.array(XDe)
```

4. Se calcula la media y la desviacion estandar de la edad

```python
print("La media de la edad es:",np.mean(XDe))
print("La desviacion estandar de la edad es:",np.std(XDe))
print("La edad minima es:",np.mean(XDe)-np.std(XDe))
print("La edad maxima es:",np.mean(XDe)+np.std(XDe))
```

5. Histograma de Frecuencias - Agrupar los datos por intervalos

```python
import matplotlib.pyplot as plt
plt.hist(XDe,bins=10)
plt.show()
```

6. Calcular el valor de pertenencia para una persona de 60 años (0,1) - Afinidad de un valor por una variable aleatoria

```python
VP=np.exp(-0.5*((50-np.mean(XDe))/np.std(XDe))**2)
print("El valor de pertenencia de la persona de 50 años es:",VP)
```
